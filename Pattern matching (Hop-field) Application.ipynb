```python
def training(x,w):
    x[x==0]=-1
    main=[]
    for i in range(0,len(x)):
        tmp=[]
        for j in range (0,len(x)):
            a=x[i]*x[j]
            tmp.append(a)
        main.append(tmp)
    main=np.array(main)
    dw=main-np.identity(len(x))
    
    w=w+dw
    return w
```
